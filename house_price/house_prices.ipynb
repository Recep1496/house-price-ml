## Görev: Eksik verileri doldurmak için bir strateji belirlemek ve uygulamak (Her bir özellik için seçilen yöntemin nedeni Notebook'ta açıklanmalıdır).

## Eksik veri doldurma planı:

Bu bölümdeki stratejimiz tamamıyla;
Türümüz sayısal ise mean() veya median() üzerinden yapılması,
türümüz kategorikde ise mode() yani en çok tekrar eden ile doldurulması,
türümüzde genel olarak çok fazla eksik varsa kısacası %80'den fazla ise sutün silme işlemi üzerinden yapılacaktır.

```python
rapor
```

```python
# 1. Sayısal ve kategorik sütunları tekrardan ayırdık
sayisal_sutunlar = dF.select_dtypes(include=['int64', 'float64']).columns
kategorik_sutunlar = dF.select_dtypes(include='object').columns

# 2. Sayısal sütunlardaki eksikleri ortalama ile doldurma işlemi yapılacak
for sutun in sayisal_sutunlar:
    if dF[sutun].isnull().sum() > 0:
        dF[sutun] = dF[sutun].fillna(dF[sutun].mean())

# 3. Kategorik sütunlardaki eksikleri mod (en sık görülen değer) ile dolduralacak.
for sutun in kategorik_sutunlar:
    if dF[sutun].isnull().sum() > 0:
        dF[sutun] = dF[sutun].fillna(dF[sutun].mode()[0])  #mod[0] almamızın sebebi en çok dönen değerlerden ilkini almak. 2tane aynı mod değeri olursa da ilkini almış olacağız.
```

 Burdan sonra yukardaki kodları tekrar yazdık. Çünkü kontrol etmek istediğimiz için

```python
from sklearn import preprocessing

# Yeni değişken ismiyle sütun adlarını al
kategorik_sutun_adlari = dF.select_dtypes(include='object').columns

# Encode işlemi
ohe = preprocessing.OneHotEncoder()
kategorik_encoded_array = ohe.fit_transform(dF[kategorik_sutun_adlari]).toarray()

# Sütun isimlerini al
kategorik_encoded_columns = ohe.get_feature_names_out(kategorik_sutun_adlari)

# DataFrame oluştur
kategorik_df = pd.DataFrame(kategorik_encoded_array, columns=kategorik_encoded_columns, index=dF.index)
```

## En iyi performansı gösteren model üzerinde GridSearchCV gibi bir yöntemle hiperparametre optimizasyonu yapmak.

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestRegressor(random_state=0)

param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

gs = GridSearchCV(estimator=rf,
                  param_grid=param_grid,
                  scoring='r2',
                  cv=10,
                  n_jobs=-1)

grid_search = gs.fit(x_train, y_train)

en_iyi_skor = grid_search.best_score_
en_iyi_parametreler = grid_search.best_params_
```

```python
en_iyi_skor
```

```python
en_iyi_parametreler
```

```python
grid_search
```

```python
import xgboost as xgb
import matplotlib.pyplot as plt

# 1. DMatrix veri setlerini oluştur
dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test, label=y_test)

# 2. Parametreleri tanımla
params = {
    'objective': 'reg:squarederror',  # regresyon için uygun
    'eval_metric': 'rmse',            # takip edilecek metrik
    'seed': 0
}

# 3. Eğitim ve doğrulama setlerini belirt
evals = [(dtrain, 'train'), (dtest, 'test')]
evals_result = {}  # sonuçları buraya yazacak

# 4. Modeli eğit (100 iterasyon)
model = xgb.train(params=params,
                  dtrain=dtrain,
                  num_boost_round=100,
                  evals=evals,
                  evals_result=evals_result,
                  verbose_eval=False)

# 5. RMSE sonuçlarını çiz
epochs = len(evals_result['train']['rmse'])
x_axis = range(epochs)

plt.figure(figsize=(10, 6))
plt.plot(x_axis, evals_result['train']['rmse'], label='Train RMSE')
plt.plot(x_axis, evals_result['test']['rmse'], label='Test RMSE')
plt.xlabel("İterasyon (Ağaç Sayısı)")
plt.ylabel("RMSE")
plt.title("XGBoost Eğitim Süreci - RMSE Değişimi")
plt.legend()
plt.grid(True)
plt.show()
```
